```python
import matplotlib.pyplot as plt
import numpy as np

# Ваши данные
results = {
    'mean': [
        7892, 27060.88, 44641.28, 70996.04,
        89023.54, 107803.19, 133940.56, 151760.1,
        178620.11, 197593.72
    ],
    'variance': [
        0, 2507129.565599996, 8243089.741599998, 10618202.2184,
        14121381.3884, 20083046.573900003, 24413497.806399994,
        28922743.710000005, 25716020.0979, 34116059.9816
    ]
}

# Рассчитываем стандартное отклонение (std = sqrt(variance))
std_dev = np.sqrt(results['variance'])

# Создаем массив для оси X (количество тиков, начиная с 1)
ticks_count = np.arange(1, len(results['mean']) + 1)

# Создаем фигуру и оси
plt.figure(figsize=(12, 6))

# Основной график средних значений
plt.plot(ticks_count, results['mean'], 'b-', label='Средний газ', linewidth=2, marker='o')

# Область отклонений (среднее ± стандартное отклонение)
plt.fill_between(ticks_count, 
                 np.array(results['mean']) - std_dev,
                 np.array(results['mean']) + std_dev,
                 color='blue', alpha=0.2, label='±1 стандартное отклонение')

# Настройки графика
plt.title('Зависимость газовых затрат от количества тиков', fontsize=14)
plt.xlabel('Количество тиков', fontsize=12)
plt.ylabel('Газ (gas units)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.xticks(ticks_count)  # Чтобы все значения по оси X были целыми числами

# Автоматическое масштабирование осей
plt.tight_layout()

# Показать график
plt.show()
```
