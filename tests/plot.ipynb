```python
std_dev = np.sqrt(results['variance'])
ticks_count = np.arange(1, len(results['mean']) + 1)
plt.figure(figsize=(12, 6))
plt.plot(ticks_count, results['mean'], 'b-', label='Средний газ', linewidth=2, marker='o')

plt.fill_between(ticks_count, 
                 np.array(results['mean']) - std_dev,
                 np.array(results['mean']) + std_dev,
                 color='blue', alpha=0.2, label='±1 стандартное отклонение')

plt.xlabel('Количество тиков', fontsize=12)
plt.ylabel('Газ (gas units)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.xticks(ticks_count) 
plt.tight_layout()

plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(ticks_count, results['mean'], 'b-', label='Средний газ', linewidth=2, marker='o')
ax1.set_xlabel('Количество тиков')
ax1.set_ylabel('Газ (gas units)', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.grid(True, linestyle='--', alpha=0.7)

ax2 = ax1.twinx()
ax2.plot(ticks_count, results['variance'], 'r--', label='Дисперсия', linewidth=1, marker='x')
ax2.set_ylabel('Дисперсия', color='r')
ax2.tick_params(axis='y', labelcolor='r')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Зависимость газа и дисперсии от количества тиков')
plt.xticks(ticks_count)
plt.tight_layout()
plt.show()
```
